```python
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
dir1 = 'output_young_pale_chubby'
```

```python
ll = os.listdir(dir1)
```

```python
ll_t = [i  for i in ll if i.endswith('.npy')]
```

```python
len(ll_t)
```

```python
norms = []
for i in ll_t:
    t = os.path.join(dir1, i)
    data = np.load(t)
    d1 = data[:,:256, :]
    d2 = data[:,256:, :]
    #print(d2.min())
    plt.figure()
    plt.imshow(d1)
    plt.figure()
    plt.imshow(d2)
    plt.figure()
    plt.imshow(d1-d2)
    plt.show()
    norms.append(np.linalg.norm(, ord=np.inf))
    break
```

```python
k = np.mean(norms)
```
